# Задание 6. Часто для качественного анализа аудитории использую подходы, основанные на сегментации. Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию. В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты, F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время. Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency, frequency и monetary для интерпретации этих кластеров. Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2) имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю, monetary от 1780 до 3560 рублей в неделю.

### В качестве метрик выберем следующие показатели:
 R - время от последней покупки пользователя до текущей даты. Текущей датой будем считать последнюю по времени дату создания заказа
 
 F - суммарное количество покупок у пользователя за всё время
 
 M - сумма покупок за всё время

### Создадим нужный нам датасет и отфильтруем по покупкам. 

```python
rfm = orders.query("order_status == ['processing', 'shipped', 'delivered']") \
    .merge(customers, how='inner', on='customer_id') \
    .merge(order_items, how='left', on='order_id')
```

```python
# посчитаем общую сумму и сгрупируем нужные нам колонки
rfm = rfm.groupby(['order_id', 'customer_unique_id', 'order_purchase_timestamp'], as_index=False) \
    .agg({'price': 'sum'})
```

```python
# считаем текущей датой последнюю по времени дату создания заказа
now = rfm.order_purchase_timestamp.max() 
```

### Создаем rfm-таблицу. Группируем по уникальному ID пользователя и агрегируем: из текущей даты вычитаем дату последней покупки клиента, считаем количество заказов на одного клиента и сумму, которую он потратил

```python
rfm = rfm.groupby('customer_unique_id', as_index=False) \
    .agg({'order_purchase_timestamp': lambda x: (now - x.max()).days,
          'order_id': lambda x: len(x.unique()),
          'price': lambda x: x.sum()}) 
```

```python
# переименуем колонки в нужные нам обозначения
rfm.rename(columns={'order_purchase_timestamp': 'recency', 
                    'order_id': 'frequency',
                    'price': 'monetary'}, inplace=True) 
```

```python
rfm.head()
```

```python
rfm.describe()
```

### Посмотрим на квантили распределения наших значений. Исходя из них логично для метрик recency и monetary выставить оценки таким образом:
для recency: 5 - наименьшее время прошедшее с последней покупки, 1 - наибольшее время соответственно

для monetory: 5 - наибольшая сумма покупок, 1 - наименьшая сумма

для frequency: мы видим что 97% - это один заказ, логично разделить оценки на два типа 1 - 1 заказ, 2 - 2 и более заказов

```python
quintiles = rfm[['recency', 'frequency', 'monetary']].quantile([.2, .4, .6, .8]).to_dict()
quintiles 
```

```python
# рассмотрим квантили frequency более подробно
rfm['frequency'].quantile([.97, .98, .999, .9999999]) 
```

### для recency: 5 - наименьшее время прошедшее с последней покупки, 1 - наибольшее время соответственно

```python
def r_score(x):
    if x <= quintiles['recency'][.2]:
        return 5
    elif x <= quintiles['recency'][.4]:
        return 4
    elif x <= quintiles['recency'][.6]:
        return 3
    elif x <= quintiles['recency'][.8]:
        return 2
    else:
        return 1
```

### для frequency: мы видим что 97% - это один заказ, логично разделить оценки на два типа 1 - 1 заказ, 2 - 2 и более заказов

```python
def f_score(x):
    if x == 1:
        return 1
    elif x > 1:
        return 2
```

### для monetory: 5 - наибольшая сумма покупок, 1 - наименьшая сумма

```python
def m_score(x):
    if x <= quintiles['monetary'][.2]:
        return 1
    elif x <= quintiles['monetary'][.4]:
        return 2
    elif x <= quintiles['monetary'][.6]:
        return 3
    elif x <= quintiles['monetary'][.8]:
        return 4
    else:
        return 5 
```

```python
rfm['r'] = rfm['recency'].apply(lambda x: r_score(x))
rfm['f'] = rfm['frequency'].apply(lambda x: f_score(x))
rfm['m'] = rfm['monetary'].apply(lambda x: m_score(x))
```

```python
# создадим объединенную колонку total_score
rfm['total_score'] = rfm['r'].map(str) + rfm['f'].map(str) + rfm['m'].map(str) 
```

```python
rfm.head()
```

```python
# словарь для замены регулярных выражений
segt_map = {
    r'[1-2][1-2]': 'hibernating',
    r'[1-2]3': 'about to sleep',
    r'[1-2][4-5]': 'promising',
    r'3[1-2]': 'at risk',
    r'33': 'need attention',
    r'3[4-5]': 'potential loyalists',
    r'[4-5][1-2]': 'can\'t loose',
    r'[4-5]3': 'loyal customers',
    r'[4-5][4-5]': 'champions'
}
```

```python
# переименуем значения в соответствии с нашим словарем.
rfm['Segment'] = rfm['r'].map(str) + rfm['m'].map(str)
rfm['Segment'] = rfm['Segment'].replace(segt_map, regex=True)
rfm.head(15)  
```

### Построим визуализацию rfm-анализа

```python
# строим график распределения клиентов по R и M оценкам
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 4))

for i, p in enumerate(['r', 'm']):
    parameters = {'r': 'recency', 'm': 'monetary'}
    y = rfm[p].value_counts().sort_index()
    x = y.index
    ax = axes[i]
    bars = ax.bar(x, y, color='silver')
    ax.set_frame_on(False)
    ax.tick_params(left=False, labelleft=False, bottom=False)
    ax.set_title('Распределение {}'.format(parameters[p]), fontsize=14)
    for bar in bars:
        value = bar.get_height()
        if value == y.max():
            bar.set_color('firebrick')
        ax.text(bar.get_x() + bar.get_width() / 2,
                value - 5,
                '{}\n({}%)'.format(int(value), int(value * 100 / y.sum())),
                ha='center',
                va='top',
                color='w')

plt.show()
```

```python
# строим распределение F в зависимости от RM рейтинга
fig, axes = plt.subplots(nrows=5, ncols=5,
                         sharex=False, sharey=True,
                         figsize=(10, 10))

r_range = range(1, 6)
m_range = range(1, 6)
for r in r_range:
    for m in m_range:
        y = rfm[(rfm['r'] == r) & (rfm['m'] == m)]['f'].value_counts().sort_index()
        x = y.index
        ax = axes[r - 1, m - 1]
        bars = ax.bar(x, y, color='silver')
        if r == 5:
            if m == 3:
                ax.set_xlabel('{}\nM'.format(m), va='top')
            else:
                ax.set_xlabel('{}\n'.format(m), va='top')
        if m == 1:
            if r == 3:
                ax.set_ylabel('R\n{}'.format(r))
            else:
                ax.set_ylabel(r)
        ax.set_frame_on(False)
        ax.tick_params(left=False, labelleft=False, bottom=False)
        ax.set_xticks(x)
        ax.set_xticklabels(x, fontsize=8)

        for bar in bars:
            value = bar.get_height()
            if value == y.max():
                bar.set_color('firebrick')
            ax.text(bar.get_x() + bar.get_width() / 2,
                    value,
                    int(value),
                    ha='center',
                    va='bottom',
                    color='k')
fig.suptitle('Распределение f для каждых r и m',
             fontsize=16)
plt.tight_layout()
plt.show()
```

### Можно проследить очевидную тенденцию по графику распределения r и m - клиенты, чаще покупающие, оставляют больше денег. Ключевая метрика - общая сумма покупок.

```python
# считаем количество клиентов в каждом сегменте
segments_counts = rfm['Segment'].value_counts().sort_values(ascending=True)

fig, ax = plt.subplots()

bars = ax.barh(range(len(segments_counts)),
               segments_counts,
               color='silver')
ax.set_frame_on(False)
ax.tick_params(left=False,
               bottom=False,
               labelbottom=False)
ax.set_yticks(range(len(segments_counts)))
ax.set_yticklabels(segments_counts.index)

for i, bar in enumerate(bars):
    value = bar.get_width()
    if segments_counts.index[i] in ['champions', 'potential loyalists', 'promising']:
        bar.set_color('firebrick')
    ax.text(value,
            bar.get_y() + bar.get_height() / 2,
            '{:,} ({:}%)'.format(int(value), int(value * 100 / segments_counts.sum())),
            va='center',
            ha='left')

plt.show()
```

### Красным отмечены пользователи с наибольшим вкладом в метрику M. Далее нам надо проанализировать клиентов, которые покупали не так давно и недавно на среднюю сумму - 'need attention' и 'loyal customers' соответственно, и попробовать перевести их в категорию 'champions' и 'promising'
